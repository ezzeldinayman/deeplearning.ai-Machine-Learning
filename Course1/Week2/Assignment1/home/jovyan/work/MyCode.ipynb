```python
def compute_cost(x, y, w, b):
    
    m = x.shape[0]
    i = 0
    total_error = 0
    for i in range(m):
        f_wb_i = w*x[i] + b
        sq_error = (f_wb_i - y[i])**2
        total_error += sq_error
    total_cost = total_error/(2*m)
    return total_cost
```

```python
initial_w = 2
initial_b = 1
cost = compute_cost(x_train, y_train, initial_w, initial_b)
print(type(cost))
print(f'cost at initial w: {cost:.3f}')
```

```python
def compute_gradient(x, y, w, b):
    
    m = x.shape[0]
    
    dj_dw = 0
    dj_db = 0
    err_w = 0
    err_b = 0
    i = 0
    for i in range(m):
        f_wb_i = w*x[i] + b
        err_w += (f_wb_i - y[i])*x[i]
        err_b += (f_wb_i - y[i])
    dj_dw = err_w/m
    dj_db = err_b/m
    return dj_dw, dj_db
```
